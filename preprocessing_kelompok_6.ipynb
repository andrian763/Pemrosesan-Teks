```python
import pandas as pd
import re
import ast

# 2. Menghapus kolom Parent yang seluruhnya NaN
data.drop(columns=['Parent'], inplace=True)

# 3. Menghapus baris dengan komentar kosong
data.dropna(subset=['Komentar'], inplace=True)

# 4. Membersihkan teks di kolom Komentar (huruf kecil, hapus angka, karakter khusus)
def clean_text(text):
    text = re.sub(r'\d+', '', text)  # Hapus angka
    text = re.sub(r'[^A-Za-z\s]', '', text)  # Hapus karakter non-alfabet
    text = re.sub(r'\s+', ' ', text)  # Menghapus spasi berlebih
    return text.lower()  # Ubah ke huruf kecil

data['Komentar'] = data['Komentar'].apply(clean_text)

# 5. Membersihkan balasan dalam list
def clean_reply_list(reply_list):
    cleaned_list = []
    for reply in reply_list:
        reply = re.sub(r'\d+', '', reply)  # Hapus angka
        reply = re.sub(r'[^A-Za-z\s,]', '', reply)  # Hapus karakter non-alfabet kecuali koma
        cleaned_list.append(reply.lower().strip())  # Ubah ke huruf kecil
    return cleaned_list

# 6. Membersihkan kolom balasan
def clean_balasan_column(balasan):
    try:
        reply_list = ast.literal_eval(balasan)  # Mengubah string ke list
        return clean_reply_list(reply_list)
    except:
        return balasan  # Kembalikan data asli jika gagal

data['balasan'] = data['balasan'].apply(clean_balasan_column)

# 7. Menampilkan hasil akhir
data
```

```python
# Fungsi untuk memproses teks
def proses_teks(teks):
    # Jika input berupa list, gabungkan menjadi string
    if isinstance(teks, list):
        teks = ' '.join([str(i) for i in teks])
    # Ubah menjadi huruf kecil
    teks_kecil = teks.lower()
    # Hapus angka
    tanpa_angka = ''.join([char for char in teks_kecil if not char.isdigit()])
    # Hapus karakter non-ASCII (misalnya, emoji, simbol khusus, karakter tidak jelas)
    tanpa_non_ascii = ''.join([char for char in tanpa_angka if ord(char) < 128])
    # Hapus karakter yang tidak diinginkan (misalnya, tanda baca)
    teks_bersih = re.sub(r'[^\w\s,]', '', tanpa_non_ascii)
    # Hapus tab dan baris baru
    teks_bersih = teks_bersih.replace('\t', ' ').replace('\n', ' ')
    # Ganti spasi ganda dengan satu spasi
    teks_bersih = re.sub(r'\s+', ' ', teks_bersih)
    # Hapus spasi di awal dan akhir
    teks_bersih = teks_bersih.strip()
    return teks_bersih

data["balasan"] = data["balasan"].apply(proses_teks)
data
```
